```python
'''
Criando um único Spark dataframe
O intuito é gravar em delta table posteriormente
'''
BUCKET_NAME = "ans-amb-cons"
FOLDER_PATH = "cons/"

# Lendo todos os CSVs da pasta e criando um unico DF
bucket_path = f"gs://{BUCKET_NAME}/{FOLDER_PATH}"
df = spark.read.csv(bucket_path, header=True, inferSchema=True, sep=";")
```

```python

df.limit(50).toPandas()
```

```python
# Caminho onde você deseja salvar os dados no formato Delta
path_to_save = "gs://processed-ans/bronze"

# Gravando o DataFrame em formato Delta
df.write.partitionBy("DT_REALIZACAO").parquet(path_to_save)
```

```python
path_to_read = "gs://processed-ans/bronze"
df_loaded = spark.read.parquet(path_to_read)
```

```python
df_loaded.limit(10).toPandas()
```
